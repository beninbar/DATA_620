Create a basic Krackhardt kite graph using libraries and data available from NetworkX and nltk. Using default data points.

```python
import networkx as nx
import nltk
from nltk.book import *
```

```python
graph1=nx.krackhardt_kite_graph()

nx.draw(graph1)
```

Let's add labels, change the color.

```python
#nx.draw(graph1, with_labels=True)
nx.draw_networkx(graph1, node_color="green", font_color="white")
```
