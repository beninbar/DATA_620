#### Create a basic Krackhardt kite graph using libraries and data available from NetworkX, using the datapoints and relationships provided.

First, try a basic Krackhardt diagram.

```python
import networkx as nx
import pandas as pd
```

```python
graph1=nx.krackhardt_kite_graph()

nx.draw(graph1)
```

Let's add labels, change the color.

```python
#nx.draw(graph1, with_labels=True)
nx.draw_networkx(graph1, node_color="green", font_color="white")
```

Add named data and assign it to the graph.

```python
relations = pd.DataFrame({'from': ['Ed', 'Ed', 'Ed', 'Beverly', 'Beverly', 'Beverly', 'Beverly', 'Andre', 'Andre', 'Andre', 
                                   'Andre', 'Carol', 'Carol', 'Carol', 'Fernando', 'Fernando', 'Fernando', 'Fernando', 
                                   'Fernando', 'Garth', 'Garth', 'Garth', 'Garth', 'Garth', 'Heather', 'Heather', 'Heather', 
                                   'Ike', 'Ike', 'Jane', 'Diane', 'Diane', 'Diane', 'Diane', 'Diane', 'Diane'],
                          'to': ['Beverly', 'Diane', 'Garth', 'Ed', 'Garth', 'Diane', 'Andre', 'Beverly', 'Diane', 
                                 'Fernando', 'Carol', 'Andre', 'Diane', 'Fernando', 'Carol', 'Andre', 'Diane', 
                                 'Garth', 'Heather', 'Ed', 'Beverly', 'Diane', 'Fernando', 'Heather', 'Fernando', 
                                 'Garth', 'Ike', 'Heather', 'Jane', 'Ike', 'Ed', 'Beverly', 'Andre', 'Carol', 
                                 'Fernando', 'Garth']})

graph2 = nx.from_pandas_edgelist(relations, 'from', 'to', create_using=nx.Graph())
nx.draw_networkx(graph2, with_labels=True, edgecolors='blue', node_color='lightgreen', node_size=500)
```
